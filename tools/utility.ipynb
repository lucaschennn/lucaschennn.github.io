### Step 1: Load Images

```python
import os
from PIL import Image

DISPLAY_SCALE = 5

category = input("daily, landscapes, or portraits")

PATH = "../public/images/portfolio/" + category + '/'
images = (sorted(os.listdir(PATH), key=lambda x: int(x.split('.')[0])))
PATHs = [(PATH + i) for i in images]

for p in PATHs:
    with Image.open(p) as img:
        img = img.resize((img.width // 10, img.height // 10))
        display(img, p + "   ^")
```

```python
from send2trash import send2trash


def utility():
    if not category:
        print("run above cell first")
        return
    while(True):
        print("CURRENTLY IN THE", category, "SUBFOLDER")
        option = input("options: [arrange], [swap], [remove], [exit]")
        if option == 'arrange':
            arrange = input("move which image where? [IM LOC] to move image IM to index LOC. [exit] to exit")
            if(arrange == "exit"):
                return

            im, loc = arrange.split(" to ")
            print(f"moving image {im} to location {loc}")
        elif option == "swap":
            swap = input("swap which two images? [IM_A IM_B] to swap images IM_A and IM_B. [exit] to exit")
            if(swap == "exit"):
                return
            im_a, im_b = swap.split()
            if(im_a not in images or im_b not in images):
                print("image not found!")
                return
            im_a_path = PATH + im_a
            im_b_path = PATH + im_b
            os.rename(im_a_path, PATH + 'temp' + im_a)
            os.rename(im_b_path, im_a_path)
            os.rename(PATH + 'temp' + im_a, im_b_path)

            print("swapping", im_a, im_b)
            
        elif option == "remove":
            to_rem = input("remove which image? [IM] to remove IM_A.jpg to the recycling bin. [exit] to exit")
            if(to_rem == "exit"):
                return
            check = input("are you sure? [Y, N]")
            if(check[0].lower() == 'y'):
                send2trash(PATH + to_rem)
        else:
            print("exiting")
            return
            
utility()
```
